## Label Encoding

```python
for col in data.select_dtypes(include=["object"]).columns:
    print(f"{col}:{data[col].unique()}")
```

```python
for col in data.select_dtypes(include=['object']).columns:
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(data[col].unique())
    data[col] = label_encoder.transform(data[col])
    print(f"{col}:{data[col].unique()}")
```

## Remove Outliers Using Z-Score

```python
def remove_outlier(data,cols,threshold=3):
    for col in cols:
        z=np.abs(stats.zscore(data[col]))
        data=data[(z<threshold) | (data[col].isnull())]
    return data
```

```python
select=['bmi']
clean=remove_outlier(data,select)
clean.shape
```

## Modeling with Hyperparameters 


```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error as MSE
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
y_pred= gbr.predict(X_test) 
y_pred
```

```python
from sklearn.metrics import r2_score
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
y_pred = gbr.predict(X_test)
rmse = r2_score(y_test, y_pred)*100
rmse
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
abr = GradientBoostingRegressor()
params = { "learning_rate": [0.01, 0.1, 1, 10],
           "n_estimators" : [20,50,100,200,150,500,600]}
grid_search = GridSearchCV(abr, param_grid=params, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train,y_train)
# Print best hyperparameters and corresponding score
print("Best hyperparameters: ", grid_search.best_params_)
```

```python
abr = GradientBoostingRegressor()
params = {"max_depth" :[5,8,15,None,10,12,13]}
grid_search = GridSearchCV(abr, param_grid=params, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train,y_train)
# Print best hyperparameters and corresponding score
print("Best hyperparameters: ", grid_search.best_params_)
```

```python
abr = GradientBoostingRegressor()
params = {"max_features" : [5,6,9,7,"auto",8],
            "min_samples_split" : [2,8,15,20,22,18] }
grid_search = GridSearchCV(abr, param_grid=params, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train,y_train)
# Print best hyperparameters and corresponding score
print("Best hyperparameters: ", grid_search.best_params_)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
abr = GradientBoostingRegressor(learning_rate=0.01, max_depth=5, max_features=5,
                          min_samples_split=20, n_estimators=500)
abr.fit(X_train,y_train)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_percentage_error
import math
y_pred = abr.predict(X_test)
mae = metrics.mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)
print('MAE is {}'.format(mae))
print('MAPE is {}'.format(mape))
print('MSE is {}'.format(mse))
print('R2 score is {}'.format(r2))
print('RMSE score is {}'.format(rmse))
```
